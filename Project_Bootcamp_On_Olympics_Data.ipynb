**2. Which country has won the most men's gold medals in Athletics over the years? Sort the results alphabetically by the player's names.**

```python
cond = (data['Gender'] == 'Men') & (data['Medal'] == 'Gold') & (data['Sport'] == 'Athletics')
data1 = data[cond]
data1
```

```python
data2 = data1.sort_values(by = 'Athlete')
data2
```

```python
data2['NOC'].value_counts()[:5]
```
