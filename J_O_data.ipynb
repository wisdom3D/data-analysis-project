# **Load data**

```python
athlete_data = pd.read_csv('/content/drive/MyDrive/jo/JeuxO/athlete_events.csv')
```

```python
region_data = pd.read_csv('/content/drive/MyDrive/jo/JeuxO/noc_regions.csv')
```

```python
data = athlete_data.merge(region_data, how = 'left', on = 'NOC')
```
