**exploration**

```python
plt.figure(figsize = [8, 6])
sns.boxplot(africa['mag'])
plt.title('magnitude')
plt.show()
```

```python
plt.figure(figsize = (8, 6))
plt.hist(data = africa, x = 'mag')
plt.title("Rpartition magnitude")
plt.xlabel("magitude")
plt.ylabel("nombre")
plt.show()
```

```python
plt.figure(figsize = (15, 10))
plt.bar(africa['NAME_LONG'].unique(), africa['NAME_LONG'].value_counts())
plt.xlabel('country', size = 50)
plt.ylabel('earthquake occurrence', size = 30)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (15, 10))
plt.bar(africa['SUBREGION'].unique(), africa['SUBREGION'].value_counts())
plt.xlabel('SUBREGION', size = 40)
plt.ylabel('earthquake occurrence', size = 30)
plt.xticks(rotation=90)
plt.title('Earthquake by region')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plt.pie(africa['SUBREGION'].value_counts(), labels = africa['SUBREGION'].value_counts().index, wedgeprops = {'width' : 0.4})
plt.title('Earthquake by region')
plt.show()
```
