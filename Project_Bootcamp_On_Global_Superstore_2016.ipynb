```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('/content/global_superstore_2016.xls')
```

```python
data.head(2)
```

```python
data.info()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data.drop(columns = ['Row ID', 'Customer ID', 'Postal Code'], inplace = True)
```

```python
data.head(1)
```

```python
if data['Sales'].str.contains('$').any():
   print("present")
```

```python
if data['Profit'].str.contains('$').any():
  print("yes")
```

```python
data['Sales'] = data['Sales'].str.replace('$', '', regex = True)
data['Sales'][:10]
```

```python
data['Sales'] = data['Sales'].str.replace(',', '', regex = True)
data['Sales'][:10]
```

```python
data['Profit'] = data['Profit'].str.replace(',', '', regex = True)
data['Profit'][:10]
```

```python
data['Profit'] = data['Profit'].str.replace('$', '', regex = True)
data['Profit'][:10]
```

```python
data['Profit'] = data['Profit'].str.replace('(', '', regex = True)
data['Profit'][:10]
```

```python
data['Profit'] = data['Profit'].str.replace(')', '', regex = True)
data['Profit'][:20]
```

```python
data['Profit'] = data['Profit'].astype(np.float64)
```
