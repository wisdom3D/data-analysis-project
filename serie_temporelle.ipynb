```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/BTC-EUR.csv', index_col = 'Date', parse_dates = True)
df.shape
```

```python
df.head(2)
```

```python
df.index
```

```python
df['Close'].plot(figsize = (9, 6))
plt.show()
```

```python
df['2021-08']['Close'].plot(figsize = (9, 6))
plt.show()
```

```python
df.loc['2021' : '2022', 'Close'].plot(figsize = (9, 6))
plt.show()
```

```python
df['2021']['Close'].resample('M').plot(figsize = (9, 6))
plt.show()
```

```python
ETH = pd.read_csv('/content/ETH-EUR.csv', index_col = 'Date', parse_dates = True)
ETH.head()
```

```python
ETH['2021']['Close'].plot(figsize = (9, 6))
plt.show()
```

```python
data = pd.merge(df, ETH, on = 'Date', how = 'inner', suffixes = ('_btc', '_eth'))
```

```python
data.head(1)
```

```python
data['2021'][['Close_btc', 'Close_eth']].plot(subplots = True, figsize = (10, 6))
plt.show()
```

```python
correlation = data[['Close_btc', 'Close_eth']].corr()
```

```python
sns.heatmap(correlation)
plt.show()
```

```python
ETH['2021-08']['Close'].diff().plot()
plt.show()
```
